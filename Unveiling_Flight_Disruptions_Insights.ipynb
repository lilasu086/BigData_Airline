```python
# Checking for Nulls in non-cancelled flights

noncancelled_flight = flights.where(col('Cancelled') == False)

null_counts = noncancelled_flight.groupby('Year').agg(*[count(when(col(c).isNull(), c)).alias(c) for c in ['CRSDepTime', 'DepTime', 'DepDelayMinutes','ArrDelayMinutes']])
null_counts.toPandas()
```

```python
# Dropping null value for DepDelayMinutes in non-cancelled Flights

flights_cleaning = flights.filter(~((col('Cancelled') == False)&(col('DepDelayMinutes').isNull())))
null_counts = flights_cleaning.where(col('Cancelled') == False).groupby('Year').agg(*[count(when(col(c).isNull(), c)).alias(c) for c in ['CRSDepTime', 'DepTime', 'DepDelayMinutes','ArrDelayMinutes']])
null_counts.toPandas()
```

```python
# For non-cancelled flights, imputing delay minutes for arrival by the value of delay minutes for departure
# Our assumption is that the Scheduled Flight Duration and actual Flight Duration are the same, which would mean that the delay in departure and arrival would be the same

from pyspark.sql.functions import coalesce

flights_cleaning = flights_cleaning.withColumn('ArrDelayMinutes', coalesce(flights_cleaning['ArrDelayMinutes'], flights_cleaning['DepDelayMinutes']))

null_counts = flights_cleaning.where(col('Cancelled') == False).groupby('Year').agg(*[count(when(col(c).isNull(), c)).alias(c) for c in ['CRSDepTime', 'DepTime', 'DepDelayMinutes','ArrDelayMinutes']])
null_counts.toPandas()
```

#### **2. Which cities experienced the highest percentage of delayed flights?**

```python
# Calculating total flights and delayed flights by city
total_flights_by_city = flights.groupBy("OriginCityName").count()
delayed_flights_by_city = flights.filter(flights.DepDelayMinutes != 0.0)\
                                    .groupBy("OriginCityName")\
                                    .count()\
                                    .withColumnRenamed('count', 'delayed_count')

# Joining the two DataFrames to get the proportion of cancelled flights
delay_proportion = total_flights_by_city.join(delayed_flights_by_city, "OriginCityName", 'left_outer')\
                                        .withColumn('delay_proportion', (col('delayed_count') / col('count')) * 100)

print("Top 10 cities with the most cancellations:")
delay_proportion.orderBy(col('delay_proportion').desc()).show(10)
```

```python
# Plotting the percent of delayed flights for above cities
delay_pd = delay_proportion.orderBy(col("delay_proportion").desc()).limit(10).toPandas()

plt.figure(figsize=(10, 6))
b=plt.bar(range(len(delay_pd)), delay_pd["delay_proportion"], color='skyblue')
plt.xlabel("Location")
plt.ylabel("Percent of Delayed Flights")
plt.title("Top 10 Cities with the Most Delayed Flights")
plt.xticks(range(len(delay_pd)), delay_pd["OriginCityName"], rotation=45, ha='right')
plt.tight_layout()
for bar, value in zip(b, delay_pd["delay_proportion"]):
    yval = bar.get_height()
    plt.text(bar.get_x() + bar.get_width()/2, yval, round(value, 2), va='bottom', ha='center')

plt.show()
```

**Insights:**

We see that a lot of regional airports with comparatively lesser number of flights are more prone to delays rather than international airports, with the exception of Dallas, Texas. We see that a little over 50% of flights from Dallas airports are delayed by some amount. This analysis provides a great opportunity to dive deeper into this airport and see if there are any particular airlines contributing to the delays or if the delays are due to some other circumstances. 

The first record here – Youngstown/Warren is very surprising as it shows that 100% of the flights from this city/ airport have been delayed. However, the total number of flights departed from this airport in 5 years is only 2. Being a primarily military airport, our assumption is that this airport is not usually used for public use in non-emergency cases.
```python
cancellation_pattern = flights.withColumn("Cancelled", F.col("Cancelled").cast("integer")).groupBy("Year", "Month").agg(
    (F.avg("Cancelled")*100).alias("Cancellation_Rate"))

total_non_cancelled_flights = flights.where(col("Cancelled")==False).count()
delay_pattern = flights.withColumn("delayed", F.when(F.col("DepDelayMinutes") != 0.0, 1).otherwise(0))\
                       .groupBy("Year", "Month")\
                       .agg(F.sum("delayed").alias("Total_Delayed"))\
                       .withColumn("Delay_Rate", F.col("Total_Delayed") / total_non_cancelled_flights * 100)


seasonal_data = cancellation_pattern.join(delay_pattern, ["Year", "Month"], 'left_outer')\
                                        .drop(delay_pattern.Total_Delayed)

seasonal_data = seasonal_data.orderBy("Year", "Month").collect()
```

## IV. Machine Learning: Classification

Our goal is to utilize a machine learning algorithm to predict the delay-time category by incorporating various features such as Origin airport, Destination airport, Air time, Airline, Day of the week, and Time of day that a flight is scheduled. This ML approach aims to forecast delay times based on fundamental flight details, offering significant benefits to travelers by providing insights into potential delays and enabling better planning for smoother travel experiences.

### Extracting Data for ML

```python
import pandas as pd

## Filtering only non-cancelled flights, to extract data for prediction
flights_nc = flights.where(col("Cancelled") == False)

pd.options.display.max_columns = None  # do not truncate the middle columns
flights_nc.limit(5).toPandas()
```

```python
# Cross-checking "Cancelled" column
flights_nc.select("Cancelled").distinct().toPandas()
```

```python
# Features: Columns to use for prediction
flights_features = flights_nc.selectExpr("Airline", "Origin", "Dest", "AirTime", "DayOfWeek", "DepTime", "DepTimeBlk", "DepDelay")
flights_features.limit(5).toPandas()
```

```python
### Creating additional features
import pyspark.sql.functions as F

## Time of Day Column
# Define conditions to categorize time of day
flights_features = flights_features.withColumn("Hour", F.hour("DepTime"))

conditions = [
    (flights_features["Hour"] >= 5) & (flights_features["Hour"] < 12),
    (flights_features["Hour"] >= 12) & (flights_features["Hour"] < 17),
    (flights_features["Hour"] >= 17) & (flights_features["Hour"] < 21),
    (flights_features["Hour"] >= 21) | (flights_features["Hour"] <= 4)
]

# Define corresponding categories
categories = ["Morning", "Afternoon", "Evening", "Night"]

# Use when() and otherwise() to categorize the time of day
flights_features = flights_features.withColumn("TimeOfDay", F.when(conditions[0], categories[0]) \
                                               .when(conditions[1], categories[1]) \
                                               .when(conditions[2], categories[2]) \
                                               .otherwise(categories[3]))
flights_features.limit(5).toPandas()
```

```python
# Corss-checking the "TimeOfDay" column
flights_features.select("TimeOfDay").distinct().toPandas()
```

```python
## Target Vairable: Delay Category Column
from pyspark.sql.functions import when

flights_features = flights_features.withColumn('DelayGroup',
                                               when(flights_features['DepDelay'] <= 0.0, 'OnTime/Early')
                                               .when((flights_features['DepDelay'] > 0.0) & (flights_features['DepDelay'] <= 15.0), 'Upto 15 mins')
                                               .when((flights_features['DepDelay'] > 15.0) & (flights_features['DepDelay'] <= 30.0), '15-30 mins')
                                               .when((flights_features['DepDelay'] > 30.0) & (flights_features['DepDelay'] <= 60.0), '30-60 mins')
                                               .when(flights_features['DepDelay'] > 60.0, 'More than 1hr')
                                               .otherwise(None))
flights_features.limit(5).toPandas()
```

```python
## PLotting the categories of the Target Variable
delayed_flights = flights_features.groupBy('DelayGroup')\
                                  .count()\
                                  .withColumnRenamed('count', 'DelayGroup_count')

delay_pd = delayed_flights.orderBy(col('DelayGroup_count').desc()).limit(10).toPandas()

plt.figure(figsize=(10, 6))
b = plt.bar(delay_pd['DelayGroup'], delay_pd['DelayGroup_count'], color='skyblue')
plt.xlabel("Delay Group")
plt.ylabel("Number of Delayed Flights")
plt.title("Delay Groups Counts")
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
for bar, value in zip(b, delay_pd['DelayGroup_count']):
    yval = bar.get_height()
    plt.text(bar.get_x() + bar.get_width()/2, yval, round(value, 2), va='bottom', ha='center')
plt.show()
```

**Note:** 

Since the 'OnTime/Eearly' flights outweigh all other delayed categories, it makes the data imbalanced across Delay Groups. For the purpose of predicting by how long a flight could possibly be delayed, we decided to keep only the delayed categories and drop/filter out the 'OnTime/Early' flights for the classification ML algorithm.

```python
# Final dataframe for classification
flights_final = flights_features.selectExpr("Airline", "Origin", "Dest", "AirTime", "DayOfWeek", "TimeOfDay", "DelayGroup")\
                                .filter(flights_features['DelayGroup'] != 'OnTime/Early')

# flights_final.limit(5).toPandas()
print("All flights:", flights_features.count())
print("Delayed flights only:", flights_final.count())
```

```python
# Corss-checking the "DelayGroup" column
flights_final.select("DelayGroup").distinct().toPandas()
```

### Building the ML Pipeline

```python
from pyspark.ml.feature import StringIndexer, OneHotEncoder, VectorAssembler
from pyspark.ml.classification import LogisticRegression, DecisionTreeClassifier, RandomForestClassifier
from pyspark.ml.evaluation import MulticlassClassificationEvaluator
from pyspark.ml import Pipeline

## Train-test split
train, test = flights_final.randomSplit([0.7, 0.3], seed = 843)

## One-Hot Encoding categorical columns
stringIndexer = StringIndexer(inputCols=['Airline', 'Origin', 'Dest', 'DayOfWeek', 'TimeOfDay', 'DelayGroup'], 
                              outputCols=['Airline_index', 'Origin_index', 'Dest_index', 'DayOfWeek_index', 'TimeOfDay_index', 'DelayGroup_index'], 
                              handleInvalid="keep")
oneHotEncoder = OneHotEncoder(inputCols=['Airline_index', 'Origin_index', 'Dest_index', 'DayOfWeek_index', 'TimeOfDay_index'], 
                              outputCols=['Airline_encoded', 'Origin_encoded', 'Dest_encoded', 'DayOfWeek_encoded', 'TimeOfDay_encoded'])

## Vector Assemblor for assembling all features in appropriate format for the estimator
assembler = VectorAssembler(inputCols=['Airline_encoded', 'Origin_encoded', 'Dest_encoded', 'DayOfWeek_encoded', 'TimeOfDay_encoded', "AirTime"], 
                            outputCol='features', handleInvalid = "skip")

## Defining Estimators

# Logistic Regression
lr = LogisticRegression().setLabelCol("DelayGroup_index").setFeaturesCol("features")

# Decision Tree Classifier
dt = DecisionTreeClassifier().setLabelCol("DelayGroup_index").setFeaturesCol("features")

# Random Forest Classifier 
rf = RandomForestClassifier().setLabelCol("DelayGroup_index").setFeaturesCol("features")

## Initializing an evaluator
evaluator = MulticlassClassificationEvaluator(labelCol="DelayGroup_index", predictionCol="prediction")
```

### Logistic Regression

```python
# Setting the stages
stages_lr = [stringIndexer, oneHotEncoder, assembler, lr]

# Adding the stages to the pipeline
pipeline_lr = Pipeline().setStages(stages_lr)

# Initializing the model
model_lr = pipeline_lr.fit(train) # fitting to the train data
preds_lr = model_lr.transform(test) # predicting on the test data

print(f"Accuracy = ",evaluator.setMetricName("accuracy").evaluate(preds_lr))
print(f"Precision = ",evaluator.setMetricName("weightedPrecision").evaluate(preds_lr))
print(f"Recall = ",evaluator.setMetricName("weightedRecall").evaluate(preds_lr))
print(f"F1 Score = ",evaluator.setMetricName("f1").evaluate(preds_lr))
```

### Decision Tree Classifier

```python
# Setting the stages
stages_dt = [stringIndexer, oneHotEncoder, assembler, dt]

# Adding the stages to the pipeline
pipeline_dt = Pipeline().setStages(stages_dt)

# Initializing the model
model_dt = pipeline_dt.fit(train) # fitting to the train data
preds_dt = model_dt.transform(test) # predicting on the test data

print(f"Accuracy = ",evaluator.setMetricName("accuracy").evaluate(preds_dt))
print(f"Precision = ",evaluator.setMetricName("weightedPrecision").evaluate(preds_dt))
print(f"Recall = ",evaluator.setMetricName("weightedRecall").evaluate(preds_dt))
print(f"F1 Score = ",evaluator.setMetricName("f1").evaluate(preds_dt))
```

### Random Forest Classifier

```python
# Setting the stages
stages_rf = [stringIndexer, oneHotEncoder, assembler, rf]

# Adding the stages to the pipeline
pipeline_rf = Pipeline().setStages(stages_rf)

# Initializing the model
model_rf = pipeline_rf.fit(train) # fitting to the train data
preds_rf = model_rf.transform(test) # predicting on the test data

print(f"Accuracy = ",evaluator.setMetricName("accuracy").evaluate(preds_rf))
print(f"Precision = ",evaluator.setMetricName("weightedPrecision").evaluate(preds_rf))
print(f"Recall = ",evaluator.setMetricName("weightedRecall").evaluate(preds_rf))
print(f"F1 Score = ",evaluator.setMetricName("f1").evaluate(preds_rf))
```

### Feature Importance: Logistic Regression

From the above results, we can see that base Logistic Regression outperforms other base models namely Decision Tree and Random Forest classifiers. However, even for Logistic Regression the evaluation metrics are pretty average. Irrespective, we have further analysed the most important features picked up by this model.

```python
# Retrieve feature metadata for both categorical and numeric features
preparedDF = stringIndexer.fit(flights_final).transform(flights_final)
preparedDF = oneHotEncoder.fit(preparedDF).transform(preparedDF)
preparedDF = assembler.transform(preparedDF)

categorical_attrs = preparedDF.schema["features"].metadata["ml_attr"]["attrs"]["binary"]
numeric_attrs = preparedDF.schema["features"].metadata["ml_attr"]["attrs"]["numeric"]

# Combine categorical and numeric attributes
feature_attrs = categorical_attrs + numeric_attrs

# Sort feature attributes by index
feature_attrs.sort(key=lambda x: x["idx"])

# Convert feature metadata to DataFrame
feature_cols = pd.DataFrame(feature_attrs)

# Set index to 'idx'
feature_cols.set_index('idx', inplace=True)
```

```python
# Retrieve coefficients from multinomial logistic regression model
coefs_array = np.array(model_lr.stages[-1].coefficientMatrix.toArray())  # Convert to numpy array
num_classes = coefs_array.shape[0]  # Number of classes

# Initialize a DataFrame to store coefficients for each class
coefs_df = pd.DataFrame(index=range(num_classes), columns=feature_cols['name'])

# Fill DataFrame with coefficients
for i in range(num_classes):
    coefs_df.loc[i] = coefs_array[i]

# Transpose DataFrame for easier interpretation
coefs_df = coefs_df.transpose()

# Rename columns to indicate class labels
coefs_df.columns = [f'coefs_class_{i}' for i in range(num_classes)]

# Sort coefficients by magnitude for each class
coefs_df = coefs_df.apply(lambda x: x.sort_values())

# Display the top coefficients for each class
coefs_df.head()
```

```python
## Plotting feature importance

# Set the figure size and create subplots
fig, axs = plt.subplots(4, 1, figsize=(10, 15), sharex=True)

# List of delay categories
delay_category = ['Upto 15 mins', '15-30 mins', '30-60 mins', 'More than 1hr']

# Iterate over each set of coefficients
for i, ax in enumerate(axs):
    # Extract coefficients and feature names for class i
    coefs = coefs_df[f'coefs_class_{i+1}']

    # Convert coefficients to numeric format
    coefs_numeric = pd.to_numeric(coefs, errors='coerce')

    # Sort coefficients by magnitude and select top 10
    top_coefs_idx = coefs_numeric.abs().nlargest(10).index
    top_coefs = coefs_numeric[top_coefs_idx]
    top_feature_names = top_coefs_idx.tolist()

    # Create a bar plot for the top 10 coefficients of class i
    ax.barh(top_feature_names, top_coefs, color='skyblue')
    # ax.set_title(f'Top 10 coefficients for coefs_class_{i}')
    ax.set_title(f'Top 10 coefficients for {delay_category[i]} Delay')
    ax.grid(True)

# Set common title and show the plot
plt.suptitle('Top 10 coefficients from the logistic regression model for each class', fontsize=16)
plt.tight_layout(rect=[0, 0.03, 1, 0.95])
plt.show()
```

**Insights:**

Based on the above feature importance plots, some of the recurring features with the highest coefficients are the origin/destination airports YNG (Youngstown-Warren Regional Airport, Ohio) and ROP (Rota International Airport, Rota Island). As discovered in the EDA section, Youngstown-Warren Regional Airport is a military and public airport that operated only 2 flights between 2018-2022 and both the flights were delayed, hence this airport is naturally picked as a strong feature to help predict flight delay. It is a similar case with Rota International Airport located in the United States Commonwealth of the Northern Mariana Islands where only a small number of flights operate, indicating lower operational efficiency. Interestingly, for more than 1 hour delay category, flights scheduled in the morning (between 5am - 11am) emerges as one of the important features in predicting this type of delay.

### Hyperparameter Tuning the Best Model: Logistic Regression

Since the above base Logistic Regression Model produced average results, we have hyperparameter tuned the model by specifying different values for a couple of hyperparameters of the model in an attempt to improve the model's performance.

```python
from pyspark.ml.tuning import ParamGridBuilder, TrainValidationSplit

# building the hyperparameter grid
params_lr = ParamGridBuilder()\
            .addGrid(lr.elasticNetParam, [0.0, 0.5, 1.0])\
            .addGrid(lr.regParam, [0.1, 0.5, 1.0, 2.0])\
            .addGrid(lr.maxIter, [10, 20, 30])\
            .addGrid(lr.threshold, [0.3, 0.5, 0.7])\
            .build()

# defining Train Validation Split to be used for hyperparameter tuning
tvs = TrainValidationSplit()\
  .setTrainRatio(0.75)\
  .setEstimatorParamMaps(params_lr)\
  .setEstimator(pipeline_lr)\
  .setEvaluator(evaluator)

tvsFitted = tvs.fit(train)  # fit the estimator

print("Best model hyperparameters:", tvsFitted.bestModel.stages[-1].extractParamMap())
print("Best model accuracy:", evaluator.setMetricName("accuracy").evaluate(tvsFitted.transform(test)))
print("Best model precision:", evaluator.setMetricName("weightedPrecision").evaluate(tvsFitted.transform(test)))
print("Best model recall:", evaluator.setMetricName("weightedRecall").evaluate(tvsFitted.transform(test)))
print("Best model F1 Score:", evaluator.setMetricName("f1").evaluate(tvsFitted.transform(test)))
```

**Insights:**

In comparison to the base model, the above results suggest that hyperparameter-tuning dropped the performance of the model rather than improving it. It's hard to tell why this might have happened and it calls for a bit more in-depth analysis of possible hyperparameters and values that could help the model. Additionally, experimenting with more number of features and using the most influential ones could also contribute to better prediction power.
